```python
import json
import pandas as pd
import pickle

# Load JSON data from file
with open("data/informatics_course_info.json", "r", encoding="utf-8") as file:
    json_data = json.load(file)

# Load Excel data
df = pd.read_excel("data/01 Programme Course options data.xlsx", sheet_name="Programme Course Options data", usecols=["Course Code", "Programme Year"])
df.dropna(inplace=True)
df["Programme Year"] = df["Programme Year"].astype(int)
year_dict = df.groupby("Course Code")["Programme Year"].apply(lambda x: tuple(x.unique())).to_dict()

print(year_dict)

# Generate the final dictionary mapping course_code to (year, semester)
yearSemester = {course["course_code"]: (year_dict.get(course["course_code"], 0), int(course["semester"][-1]) if (course["semester"][-1] if course["semester"]!='' else 'x').isdigit() else 0) for course in json_data}

# Print or use the dictionary
print(yearSemester)

with open('data/yearSemester.pkl', 'wb') as f:
    pickle.dump(yearSemester, f)
```
